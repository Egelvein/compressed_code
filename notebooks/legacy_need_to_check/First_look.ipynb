## Checking on death tokens

```python
def find_dead_tokens(tokenizer):
    vocab = tokenizer.get_vocab()
    unused_tokens = []
    for token, token_id in vocab.items():
        decoded = tokenizer.decode([token_id])
        if not decoded.strip():
            unused_tokens.append((token, token_id))
    return unused_tokens
```

```python
dead_tokens_by_tokenizer = {}

for name, tokenizer in tokenizers_list:
    print(name)
    dead_tokens = find_dead_tokens(tokenizer)
    dead_tokens_by_tokenizer[name] = dead_tokens
```

```python
dead_qwen_coder = len(dead_tokens_by_tokenizer['qwen_coder'])
dead_qwen = len(dead_tokens_by_tokenizer['qwen'])
dead_athene = len(dead_tokens_by_tokenizer['athene'])
dead_deepseek = len(dead_tokens_by_tokenizer['deepseek'])
dead_deepseek_r1 = len(dead_tokens_by_tokenizer['deepseek_r1'])
dead_llama = len(dead_tokens_by_tokenizer['llama'])
```

```python
languages = [
    'Python', 'Java', 'C++', 'Go', 'React', 'JavaScript', 'Rust', 'PHP'
]

prefix_to_language = {
    'py': 'Python',
    'ja': 'Java',
    'cpp': 'C++',
    'go': 'Go',
    're': 'React',
    'js': 'JavaScript',
    'ru': 'Rust',
    'ph': 'PHP'
}

language_tokens = {lang: [] for lang in prefix_to_language.values()}
language_counts = {lang: [] for lang in prefix_to_language.values()}
```

## Qwen-Coder

```python
#         

print("""The larger the number, the less frequently the given word was found in the training dataset.
If the number is less than 0 - this word wasn't in the training dataset.""")

negative_tokens = {
    'qwen_tokenizer': {lang: [] for lang in prefix_to_language.values()}
}

for token, count in qwen_coder_spec_toks.items():
    for prefix, language in prefix_to_language.items():
        if token.startswith(f"{prefix}_"):
            language_tokens[language].append(token)
            language_counts[language].append(count)
            if count == -5000:
                negative_tokens['qwen_tokenizer'][language].append(token)
            break

fig, ax = plt.subplots(len(prefix_to_language), 1, figsize=(10, len(prefix_to_language) * 5))

for i, language in enumerate(prefix_to_language.values()):
    if language_tokens[language]:
        ax[i].bar(language_tokens[language], language_counts[language])
        ax[i].set_title(f"Token Frequency in {language}")
        ax[i].set_xlabel('Token')
        ax[i].set_ylabel('Rank in tokenizer')
        ax[i].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
